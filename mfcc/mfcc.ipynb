```python
import librosa
audio_path = './REC20200222223320.mp3.wav.wav'
x_1 , sr_1 = librosa.load(audio_path,sr=None)
print(x_1, sr_1)
x_1_16 , sr_1_16 = librosa.load(audio_path,sr=16000)
print(x_1, sr_1_16)
x_1_8 , sr_1_8 = librosa.load(audio_path,sr=8000)
print(x_1, sr_1_8)

#display waveform
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x_1, sr=sr_1)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x_1, sr=sr_1_16)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x_1, sr=sr_1_8)

#display Spectrogram
X = librosa.stft(x_1)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr_1, x_axis='time', y_axis='hz') 
#If to pring log of frequencies  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()

#display Spectrogram
X = librosa.stft(x_1_16)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr_1_16, x_axis='time', y_axis='hz') 
#If to pring log of frequencies  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()

#display Spectrogram
X = librosa.stft(x_1_8)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr_1_8, x_axis='time', y_axis='hz') 
#If to pring log of frequencies  
#librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()




```

```python
mfccs = librosa.feature.mfcc(x_1, sr=sr_1,n_mfcc=13)
print(mfccs.shape)
#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr_1, x_axis='time',cmap='gray_r')
```
